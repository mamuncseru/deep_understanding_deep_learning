## import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

from IPython import display
display.set_matplotlib_formats('svg')
```

```python
net_noseed = nn.Sequential(
    nn.Linear(2, 8), # input layer
    nn.Linear(8, 8), # hidden layer
    nn.Linear(8, 1), # output layer 
)
```

```python
torch.manual_seed(1)
net_rs1a = nn.Sequential(
    nn.Linear(2, 8), # input layer
    nn.Linear(8, 8), # hidden layer
    nn.Linear(8, 1), # output layer 
)
```

```python
torch.manual_seed(2)
net_rs2 = nn.Sequential(
    nn.Linear(2, 8), # input layer
    nn.Linear(8, 8), # hidden layer
    nn.Linear(8, 1), # output layer 
)
```

```python
torch.manual_seed(1)
net_rs1b = nn.Sequential(
    nn.Linear(2, 8), # input layer
    nn.Linear(8, 8), # hidden layer
    nn.Linear(8, 1), # output layer 
)
```

```python
np.concatenate()
```

```python
noseed_weights = np.zeros((1))
for p in net_noseed.named_parameters():
    if 'weight' in p[0]:
        noseed_weights = np.concatenate((noseed_weights, p[1].data.numpy().flatten()), axis=0)

noseed_weights
```

```python
rs1a_weights = np.zeros((1))
for p in net_rs1a.named_parameters():
    if 'weight' in p[0]:
        rs1a_weights = np.concatenate((rs1a_weights, p[1].data.numpy().flatten()), axis=0)

rs1a_weights
```

```python
rs2_weights = np.zeros((1))
for p in net_rs2.named_parameters():
    if 'weight' in p[0]:
        rs2_weights = np.concatenate((rs2_weights, p[1].data.numpy().flatten()), axis=0)

rs2_weights
```

```python
rs1b_weights = np.zeros((1))
for p in net_rs1b.named_parameters():
    if 'weight' in p[0]:
        rs1b_weights = np.concatenate((rs1b_weights, p[1].data.numpy().flatten()), axis=0)

rs1b_weights
```

```python
plt.figure(figsize=(15, 5))
plt.plot(noseed_weights, 'ro', label = 'no seed')
plt.plot(rs1a_weights, 'ks', label = 'rs1a')
plt.plot(rs2_weights, 'c^', label = 'rs2')
plt.plot(rs1b_weights, 'g+', label = 'rs1b')
plt.legend()
plt.xlabel('Vectorized weight index')
plt.ylabel('Weight value')
plt.show()
```
